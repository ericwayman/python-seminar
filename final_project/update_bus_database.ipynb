```python
#imports
import pandas as pd
import sqlite3
import pandas.io.sql as sql
import numpy as np
```

```python
#make sql connection
connection = sqlite3.connect('bus_arrival.db')
#import the database as a pandas df
df= sql.read_frame(sql='SELECT trip_id,bus_id,arrival_time,stop_id,demand, delay FROM actual_arrival', con=connection)

#convert arrival time to a date time object
df['arrival_time'] = pd.to_datetime(df['arrival_time'])
```

```python
def add_prev_stop_delay(df,lag):
    """inputs: df: a data frame
                lag: an int that represent how many stops forward are being predicted.  
        output: a new dataframe
        returns a df that is a modification of the input dataframe by adding a column k_prev_stop_delay, where k=lag.  This column computes the lag on
        the trip for the stop k stops previous to the stop of the current row.  When this data doesn't exist, the entry is NaN.
    
    """
    num_rows = len(df.index)
    #sort the dataframe in sequential order grouped by trip
    df = df.sort(['trip_id','bus_id','arrival_time'], ascending=True)
    #update the index
    df.index = pd.Series(np.arange(num_rows))
    #add the new column
    df['%i_prev_stop_delay' %lag] = df['delay'].shift(lag)
    #eliminate prev_stop_delay values that don't correspond to previous stops
    for i in np.arange(lag,num_rows):
        #check if shifted column belongs to the same trip and bus
        if (df['trip_id'][i] != df['trip_id'][i-lag] or df['bus_id'][i] != df['bus_id'][i-lag] ):
            df['%i_prev_stop_delay' %lag][i] = float('NaN')
        #check if the shifted column corresponds to a bus stop k stops previous (i.e. shifted_stop = current_stop + lag)
        elif (df['stop_id'][i] != df['stop_id'][i-lag]-lag):
            df['%i_prev_stop_delay' %lag][i]=float('NaN')
    
    return df
    
```

```python
def add_pred_delay(df,lag):
    """inputs: df: a data frame
            lag: an int that represent how many stops forward are being predicted.  
    output: a new dataframe
    returns a df that is a modification of the input dataframe by adding a column k_pred_delay, where k=lag.  
    This column finds the delay at the stop for the last bus to arrive at the future stop: stop x - lag before the current time
    which is the arrival time of the current bus at stop x
    k_pred_delay represents the delay of the predecessor k stops ahead
    stop_id ranges from 37 (last stop) to 46 (first stop)
    """
    num_rows = len(df.index)
    #sort the dataframe
    df = df.sort(['arrival_time'], ascending=True)
    #update the index
    df.index = pd.Series(np.arange(num_rows))
    
    #initialize the pred_delay_list
    pred_delay_list = num_rows*[float('NaN')]
    for x in np.arange(num_rows):
        #Check to make sure there are still lag more steps left in the route
        if (df['stop_id'][x]-lag  < 37):
            pass
        else:
            #find the delay for the last bus to arrive for the future stop: x - lag
            pred_df = df[(df.stop_id==df['stop_id'][x]-lag)& (df.arrival_time < df.arrival_time[x])]
            if (len(pred_df.index)==0):
                pass
            else:
                #set the delay for the last arrival at the stop before the current time
                pred_delay_list[x] = list(pred_df['delay'])[-1]
    
    #add the new list as a new column to the dataframe
    df['%i_pred_delay' %lag] = pd.Series(pred_delay_list)
    return df
```

```python
# add prev stop delays to df
for lag in np.arange(1,8):
    df = add_prev_stop_delay(df,lag)
```
