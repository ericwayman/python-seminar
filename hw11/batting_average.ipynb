```python
#imports
import bat_avg_model
import pymc
import pandas as pd
import matplotlib as plt
import numpy as np
%pylab inline

#read in the data as a data frame
april_df = pd.read_table('./hw_11_data/laa_2011_april.txt',sep='\t')
```

```python
#compute the MLE of the batting average for each player.  For X~binomial(N,p),
#the MLE for p is justX/N
#so our MLE for the batting average is just the observed batting average
MLE_bat_avg = april_df['AVG']
print MLE_bat_avg
```

```python
#run MCMC on the player batting data from the month of april.  
#The number of hits is assumed to follow a binomial model
#assumes a file called bat_avg_model.py is in the PWD
M = pymc.MCMC(bat_avg_model)
```

```python
N_samp = 20000
N_burn = 1000
M.sample(N_samp, burn=N_burn, thin=2)
```

```python
#compute the posterior means of the batting averages for the 13 players
post_ba_means = M.stats()['ba']['mean']
print list(post_ba_means)
```

```python
#plot the traces for the samples from the posterior batting averages for the first 3 players
fig, (ax1, ax2,ax3) = plt.subplots(1, 3, sharey=True)
fig.set_size_inches(10,5)
fig.suptitle('Trace plots for the first three batters')
ax1.plot(M.ba.trace[:,0])
ax2.plot(M.ba.trace[:,1])
ax3.plot(M.ba.trace[:,2])
ax1.set_title('%s'%april_df['Player'][0])
ax2.set_title('%s'%april_df['Player'][1])
ax3.set_title('%s'%april_df['Player'][2])
```

```python
full_df = pd.read_table('./hw_11_data/laa_2011_full.txt')
season_ba = full_df['AVG']
print season_ba
```

```python
#Find the 95% central posterior CI for each players batting average coefficient.
alpha = 0.05
num_players = len(april_df.index)
#list recording whether or not predicted value falls in the confidence intervals
in_CI = []
#the upper and lower bounds for the 95% confidence intervals
upper_bound = []
lower_bound=[]

print "Posterior for the 95 percent confidence intervals for each player"
for i in range(num_players):
    N = len(M.ba.trace[:,i])
    sorted_ba_posterior =  np.sort(M.ba.trace[:,i])
    print "%s: (%f, %f)" % (april_df['Player'][i],sorted_ba_posterior[np.ceil(N*alpha/2.)], sorted_ba_posterior[np.ceil(N*(1-alpha/2.))])
    #check if the player's full season batting average falls in the 95 percent confidence interval
    in_CI.append((sorted_ba_posterior[np.ceil(N*alpha/2.)] < season_ba[i]) & (season_ba[i]<sorted_ba_posterior[np.ceil(N*(1-alpha/2.))]))
    upper_bound.append(sorted_ba_posterior[np.ceil(N*alpha/2.)])
    lower_bound.append(sorted_ba_posterior[np.ceil(N*(1-alpha/2.))])
```

```python
in_CI
```

```python
print("so %i of the %i players had their season batting averages fall within the 95 percent confidence interval"
%(in_CI.count(True),num_players))
```

```python
#make plots
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ln1 = ax.plot(season_ba,'x',color='red',label='Season Batting Average')
ln2 = ax.plot(MLE_bat_avg,'o', color ='blue',label='MLE batting average')
plt.ylabel("Batting Average",fontsize = 16)

#making the legend
lns = ln1+ln2
labels = [l.get_label() for l in lns]
leg = ax.legend(lns, labels,loc=1, prop={'size':10})
```

```python
#make plots
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)
ln1 = ax.plot(season_ba,'x',color='red',label='Season Batting Average')
ln2 = ax.errorbar(np.arange(num_players),post_ba_means,fmt='o',yerr=[post_ba_means-lower_bound,upper_bound-post_ba_means],label='Posterior Mean Batting Average')
plt.ylabel("Batting Average",fontsize = 16)

#making the legend
#lns = ln1+ln2
#labels = [l.get_label() for l in lns]
#leg = ax.legend(lns, labels,loc=1, prop={'size':10})
```
