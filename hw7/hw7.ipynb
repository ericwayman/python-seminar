```python
import pandas as pd
#read in the json to a dataframe
result = pd.read_json('./closed.json')

#3 remove duplicate rows by id

#find the duplicated indices
duplicated_idxs = result.duplicated(cols='id')

#keep only non duplicated indices for the data frame
result = result.ix[duplicated_idxs == False]


#1 Make a DataFrame with one row per issue 
#with the following columns extracted from the issue data:
#ntitle, created_at, labels, closed_at, user, id

df = result[['title','created_at','labels','closed_at','user','id']]

#2 convert the dicts in the user columns to strings representing the user names

df['user'] = [x['login'] for x in df['user']]
df.head()

#4 don't need to convert created_at and closed_at columns to datetime because
#they were already imported as datetime objects
```

```python
df
```

```python
#5 Plot number of issues and number of distinct users by month

#set 'created_at' as the dataframe index.  
#drop = False makes sure we keep 'created_at' as a column
df.set_index('created_at', inplace=True, drop = False)

#define a function to count the number of distinct users
def user_count(x):
    return len(set(x))

#make series containing number of distinct users per month
user_results = df['user'].resample('M',how =user_count)

#check to see if executed properly
user_results[:10]
```

```python
% pylab inline
user_results.plot()
```

```python
#compute a series containing the number of issues per month
issues_results = df['created_at'].resample('M', how=['count'])
issues_results[:10]
```

```python
issues_results.plot()
```

```python
#6 make a table computing the mean number of days to resolve an issue

#series of total seconds required to resolve each issue
resolve_times = [double((df['closed_at'][i]-df['created_at'][i]).total_seconds())/(60.0*60.0*24.0) for i in range(len(df['closed_at']))]
table = pd.DataFrame({'created_at':df['created_at'],'resolve_time':resolve_times})

#resample the resolve times to compute a series containing the mean resolve_time for each month
mean_resolve_time = table['resolve_time'].resample('M', how=['mean'])
```

```python
#make the table with mean number number of days to solve an issue and total number of issues per month
closed_issues_table = pd.DataFrame({'mean_resolve_time_in_days':mean_resolve_time['mean'],'num_issues':issues_results['count']})
closed_issues_table
```

```python
#7  Make a DataFrame containing all the comments for all of the issues.
#result['comments'][i] is an array whose single element is a dict, so result['comments'][i][0] is of type dict.  
#extract these dicts into a list, if the comments for an issues is not empty
comment_data = [dict(result['comments'][i][0],**{'id':result['id'][i]}) for i in result.index if result['comments'][i] != []]
comment_df = pd.DataFrame(comment_data)
comment_df.head()
```

```python
#convert the to 'created' column to a datetime object
comment_df['created']=(comment_df['created'].map(int)*1000000).map(pd.to_datetime)
```

```python
comment_df.set_index('created', inplace=True)
comment_df.head()
```

```python
#define a function to count the number of distinct users
def user_count(x):
    return len(set(x))

#define function to find chattiest use
from operator import itemgetter
from collections import Counter

def chattiest(my_list):
    if len(my_list)==0: 
        return  None
    author_count = Counter(my_list)
    return max(author_count.iteritems(), key=itemgetter(1))[0]

def percent_comments(my_list):
    if len(my_list) ==0:
        return 0.0
    author_count = Counter(my_list)
    numerator = double(max(author_count.iteritems(), key=itemgetter(1))[1])
    denominator = double(sum(author_count.values()))
    return 100*(numerator/denominator)
```

```python
#compute a table sumarizing for each month:
#number of issue comments, "chattiest" user, percent of comments by chattiest, and number of commentors
chat_df = comment_df['author'].resample('M',how =['count',user_count,chattiest,percent_comments])
chat_df
```

```python
#create the helper labels table
label_helpers_data=[]
for i in result.index:
    if result['labels'][i]==[]:
        label_helpers_data.append((result['id'][i],None))
    else:
        for j in range(len(result['labels'][i])):
            label_helpers_data.append((result['id'][i],result['labels'][i][j]))
label_df = pd.DataFrame(label_helpers_data,columns = ['id','label'])
label_df.head()    
```
